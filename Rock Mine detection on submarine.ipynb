```python
datas.head(2)
```

```python
datas.value_counts() 
datas.groupby('Result').mean()
```

```python
#Segrigating Datas and Result
data=datas.copy()
x=data.drop('Result', axis=1)
y=data['Result']
```

```python
data.describe()
```

```python
profile=ProfileReport(datas)
profile.to_widgets()
```

```python
profile.to_file('report.html')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Applying suitable Model and assigning Train and Test data
x_train, x_test, y_train, y_test= train_test_split(x,y, train_size=0.7, random_state=10, shuffle= True, stratify=y)
model=LogisticRegression()
model.fit(x_train, y_train)
```

```python
#Training Data and getting accuracy
x_train_prediction=model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print(training_data_accuracy)
```

```python
#Testing Data and checking validity
x_test_prediction=model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction, y_test)
print(testing_data_accuracy)
```

Our Model is 80.6% train accuracy and 77.7% test accuracy
